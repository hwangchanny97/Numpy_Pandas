# **1-2. 라인 스타일(Line Style)**
|문자  ||스타일    |
|-----||:-------|
| `-`   ||solid  |
| `--`  ||dashed |
|`-.`  ||dashdot |
| `:`   ||dotted  |

```python
# 라인 스타일 지정 1

plt.plot(np.random.rand(50).cumsum(), linestyle='solid')
plt.plot(np.random.rand(50).cumsum(), linestyle='dashed')
plt.plot(np.random.rand(50).cumsum(), linestyle='dashdot')
plt.plot(np.random.rand(50).cumsum(), linestyle='dotted');
```

```python
# 라인 스타일 지정2

plt.plot(np.random.rand(50).cumsum(), linestyle='-')
plt.plot(np.random.rand(50).cumsum(), linestyle='--')
plt.plot(np.random.rand(50).cumsum(), linestyle='-.')
plt.plot(np.random.rand(50).cumsum(), linestyle=':');
```

# **1-5. 플롯 레이블(Plot Label)**

```python
# title, xlabel, ylabel

plt.plot(np.random.randint(0, 51, 50))
plt.title('exam score')
plt.xlabel('student number')
plt.ylabel('score')
```

```python
# plot에 label 추가

plt.plot(np.random.rand(50).cumsum(), label='StudentA', linestyle=':', color='r')
plt.plot(np.random.rand(50).cumsum(), label='StudentB', linestyle='-', color='g')
plt.plot(np.random.rand(50).cumsum(), label='StudentC', linestyle='--', color='b')

plt.title('exam score')
plt.xlabel('student number')
plt.ylabel('score')
plt.legend();
```

```python
# 범례(legend) 추가 

plt.plot(np.random.rand(50).cumsum(), 'r-',label='Student A')
plt.plot(np.random.rand(50).cumsum(), 'g--',label='Student B')
plt.plot(np.random.rand(50).cumsum(), 'b-.',label='Student C')

plt.title('exam score')
plt.xlabel('exam count')
plt.ylabel('score')
plt.legend();
```

# **1-6. 막대 플롯(Bar Plot)**

```python
# bar, xticks

height = [np.random.rand() * 10 * i for i in range(1,6)]
names = ['Student A', 'Student B', 'Student C', 'Student D', 'Student E']
bar_cnt = np.arange(len(names))

plt.bar(bar_cnt, height)
plt.xticks(bar_cnt, names, fontweight='bold')

plt.xlabel('student name')
plt.ylabel('score')
```

```python
# barh, yticks

height = [np.random.rand() * i for i in range(1, 6)]
names = ['Student A', 'Student B', 'Student C', 'Student D', 'Student E']
bar_cnt = np.arange(len(names))

plt.barh(bar_cnt, height)

plt.yticks(bar_cnt, names, fontweight='bold')

plt.ylabel('student name')
plt.xlabel('score');
```

```python
# 누적 막대

bar1 = np.array([1,2,3,4,5])
bar2 = np.array([10,11,12,13,14])
bar3 = [15,16,17,18,19]

bar_sum = bar1 + bar2

bar_cnt = np.arange(5)
names = ['Student A', 'Student B', 'Student C', 'Student D', 'Student E']

plt.barh(bar_cnt, bar1, color='darkblue')
plt.barh(bar_cnt, bar2, left=bar1, color='skyblue')
plt.barh(bar_cnt, bar3, left=bar_sum, color='royalblue')

plt.yticks(bar_cnt, names, fontweight='bold')
plt.ylabel('Student name')
plt.xlabel('Score')
```

```python
bar1 = np.array([1, 2, 3, 4, 5])
bar2 = np.array([10, 11, 12, 13, 14])
bar3 = [15, 16, 17, 18, 19]
bar_sum = bar1 + bar2

bar_cnt = np.arange(5)
names = ['Student A', 'Student B', 'Student C', 'Student D', 'Student E']

plt.bar(bar_cnt, bar1, color='darkblue')
plt.bar(bar_cnt, bar2, bottom=bar1, color='skyblue')
plt.bar(bar_cnt, bar3, bottom=bar_sum, color='royalblue')

plt.xticks(bar_cnt, names, fontweight='bold')
plt.xlabel('student name')
plt.ylabel('score');
```
