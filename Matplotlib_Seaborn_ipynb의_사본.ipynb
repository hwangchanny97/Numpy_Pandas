# **1-2. 라인 스타일(Line Style)**
|문자  ||스타일    |
|-----||:-------|
| `-`   ||solid  |
| `--`  ||dashed |
|`-.`  ||dashdot |
| `:`   ||dotted  |

```python
# 라인 스타일 지정 1

plt.plot(np.random.rand(50).cumsum(), linestyle='solid')
plt.plot(np.random.rand(50).cumsum(), linestyle='dashed')
plt.plot(np.random.rand(50).cumsum(), linestyle='dashdot')
plt.plot(np.random.rand(50).cumsum(), linestyle='dotted');
```

```python
# 라인 스타일 지정2

plt.plot(np.random.rand(50).cumsum(), linestyle='-')
plt.plot(np.random.rand(50).cumsum(), linestyle='--')
plt.plot(np.random.rand(50).cumsum(), linestyle='-.')
plt.plot(np.random.rand(50).cumsum(), linestyle=':');
```

# **1-5. 플롯 레이블(Plot Label)**

```python
# title, xlabel, ylabel

plt.plot(np.random.randint(0, 51, 50))
plt.title('exam score')
plt.xlabel('student number')
plt.ylabel('score')
```

```python
# plot에 label 추가

plt.plot(np.random.rand(50).cumsum(), label='StudentA', linestyle=':', color='r')
plt.plot(np.random.rand(50).cumsum(), label='StudentB', linestyle='-', color='g')
plt.plot(np.random.rand(50).cumsum(), label='StudentC', linestyle='--', color='b')

plt.title('exam score')
plt.xlabel('student number')
plt.ylabel('score')
plt.legend();
```

```python
# 범례(legend) 추가 

plt.plot(np.random.rand(50).cumsum(), 'r-',label='Student A')
plt.plot(np.random.rand(50).cumsum(), 'g--',label='Student B')
plt.plot(np.random.rand(50).cumsum(), 'b-.',label='Student C')

plt.title('exam score')
plt.xlabel('exam count')
plt.ylabel('score')
plt.legend();
```

# **1-6. 막대 플롯(Bar Plot)**

```python
# bar, xticks

height = [np.random.rand() * 10 * i for i in range(1,6)]
names = ['Student A', 'Student B', 'Student C', 'Student D', 'Student E']
bar_cnt = np.arange(len(names))

plt.bar(bar_cnt, height)
plt.xticks(bar_cnt, names, fontweight='bold')

plt.xlabel('student name')
plt.ylabel('score')
```

```python
# barh, yticks

height = [np.random.rand() * i for i in range(1, 6)]
names = ['Student A', 'Student B', 'Student C', 'Student D', 'Student E']
bar_cnt = np.arange(len(names))

plt.barh(bar_cnt, height)

plt.yticks(bar_cnt, names, fontweight='bold')

plt.ylabel('student name')
plt.xlabel('score');
```

```python
# 누적 막대

bar1 = np.array([1,2,3,4,5])
bar2 = np.array([10,11,12,13,14])
bar3 = [15,16,17,18,19]

bar_sum = bar1 + bar2

bar_cnt = np.arange(5)
names = ['Student A', 'Student B', 'Student C', 'Student D', 'Student E']

plt.barh(bar_cnt, bar1, color='darkblue')
plt.barh(bar_cnt, bar2, left=bar1, color='skyblue')
plt.barh(bar_cnt, bar3, left=bar_sum, color='royalblue')

plt.yticks(bar_cnt, names, fontweight='bold')
plt.ylabel('Student name')
plt.xlabel('Score')
```

```python
bar1 = np.array([1, 2, 3, 4, 5])
bar2 = np.array([10, 11, 12, 13, 14])
bar3 = [15, 16, 17, 18, 19]
bar_sum = bar1 + bar2

bar_cnt = np.arange(5)
names = ['Student A', 'Student B', 'Student C', 'Student D', 'Student E']

plt.bar(bar_cnt, bar1, color='darkblue')
plt.bar(bar_cnt, bar2, bottom=bar1, color='skyblue')
plt.bar(bar_cnt, bar3, bottom=bar_sum, color='royalblue')

plt.xticks(bar_cnt, names, fontweight='bold')
plt.xlabel('student name')
plt.ylabel('score');
```

# **2. Seaborn이란?**
 - Matplotlib를 기반으로 하느 Python 데이터 시각화 라이브러리
 - 매력적이고 유익한 통계 그래픽을 그리기 위한 고급 인터페이스를 제공
 - seaborn 웹페이지: http://seaborn.pydata.org/

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# seaborn 임포트
import seaborn as sns

# style 설정
sns.set(style='darkgrid')
```

```python
# tips 데이터셋 불러오기

tips = sns.load_dataset('tips')
tips
```

# **2-1. 관계 플롯(Rel Plot)**

- seaborn.relplot: https://seaborn.pydata.org/generated/seaborn.relplot.html

```python
# relplot

sns.relplot(x='total_bill', y='tip', data=tips);
```

```python
# tips unique

tips['time'].unique()
```

```python
# hue 색상

sns.relplot(x='total_bill', y='tip', hue='time', data=tips);
```

```python
# style 지정

sns.relplot(x='total_bill', y='tip', style='smoker',data=tips);
```

```python
# style과 hue

sns.relplot(x='total_bill', y='tip', style='smoker', hue='smoker',data=tips);
```

```python
# palette

sns.relplot(x='total_bill', y='tip', hue='size', palette='viridis',data=tips);
```

```python
# size

sns.relplot(x='total_bill', y='tip', hue='size', size='size', palette='viridis',data=tips);
```

# **2-2. 카테고리 플롯(Cat Plot)**
- seaborn.catplot: https://seaborn.pydata.org/generated/seaborn.catplot.html

```python
# catplot

sns.catplot(x='day', y='total_bill', data=tips);
```

```python
# kind로 종류 변경

sns.catplot(x='day', y='total_bill', kind='violin',data=tips)
```

```python
# hue

sns.catplot(x='day', y='total_bill', kind='swarm', hue='sex',data=tips)
```

```python
# hue 2

sns.catplot(x='sex', y='total_bill', kind='swarm', hue='smoker',data=tips)
```

```python
# split

sns.catplot(x='day', y='total_bill', kind='violin', hue='sex', split=True, data=tips)
```

```python
# titanic 데이터셋 가져오기

tic = sns.load_dataset('titanic')
tic
```

```python
# titanic unique

tic['embark_town'].unique()
```

```python
# 탑승지점 별 나이

sns.catplot(x='embark_town', y='age', data=tic);
```

```python
# count

tic.groupby('embark_town').count()
```

```python
# 성별에 따른 생존률

sns.catplot(x='sex', y='survived', kind='bar', data=tic)
```

```python
# 성별에 따른 생존률 2

sns.catplot(x='sex', y='survived',hue='class' ,kind='bar', data=tic)
```

# **2-3. 페어 플롯(Pair Plot)**
 - seaborn.pairplot: https://seaborn.pydata.org/generated/seaborn.pairplot.html

```python
# penguins 데이터셋 불러오기

pgs = sns.load_dataset('penguins')
pgs
```

```python
# penguins unique

print(pgs['species'].unique())
print(pgs['island'].unique())
```

- 펭귄 종류

penguin species.PNG

- 펭귄 신체

penguin body.PNG

```python
# pairplot

sns.pairplot(pgs);
```

```python
# hue

sns.pairplot(pgs, hue='species');
```

```python
# palette

sns.pairplot(pgs, hue='species', palette='viridis');
```

```python
# markers

sns.pairplot(pgs, hue='species', palette='viridis', markers=['o','s','d']);
```

```python
# x_vars, y_vars

sns.pairplot(pgs, hue='species', palette='viridis', markers=['o','s','d'],
             x_vars=['flipper_length_mm', 'body_mass_g', 'bill_depth_mm'], y_vars=['bill_length_mm']);
```
